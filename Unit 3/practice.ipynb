```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn import metrics

data = pd.read_csv("Social_Network_Ads.csv")
print(data)
print(data.head())
print(data.shape)

x = data.iloc[:, [2,3]]
y = data.iloc[:, 4]
print(x.head())
print(y.head())

X_train, X_test, Y_train, Y_test = train_test_split(x, y, test_size=0.3, random_state=42)

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

classifier = SVC(kernel='linear', random_state=42)
classifier.fit(X_train, Y_train)

Y_pred = classifier.predict(X_test)
print("Prediction: \n", Y_pred)

print("Accuracy: ", metrics.accuracy_score(Y_test, Y_pred))

plt.figure(figsize=(8,4))
plt.scatter(X_test[:, 0], X_test[:, 1], c=Y_test, cmap='viridis', marker='D', s=60, label='Data')

w = classifier.coef_[0]
a = -w[0]/w[1]
xx = np.linspace(X_test[:, 0].min(), X_test[:, 0].max())
yy = a * xx - (classifier.intercept_[0] / w[1])
plt.plot(xx, yy, 'r--', label="Decision Boundary")

plt.title("HAHAHAHA")
plt.xlabel("AGE")
plt.ylabel("SALARY")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

x = np.array([1.5, 3.0, 2.0, 2.5, 3.0, 4.0]).reshape(-1, 1)
y = np.array([50, 84, 60, 65, 80, 90]) 

model = LinearRegression().fit(x, y)

new_study_hours = np.array([[3.5]])
predicted_score = model.predict(new_study_hours)
print(f'Predicted Exam Score for 3.5 Hours Study : {predicted_score[0]:.2f}')

plt.figure(figsize=(10, 6))
plt.scatter(x, y, color='red', marker='o', s=70, label='Actual Data')
plt.plot(x, model.predict(x), color='seagreen', linestyle='dashed', label='Regression Line')
plt.scatter(new_study_hours, predicted_score, color='cyan', marker='^', s=100, label=f'Predicted Exam Score : {predicted_score[0]:.2f}')

plt.title("Linear Regression - List/Array")
plt.xlabel("Study Hours")
plt.ylabel("Exam Score")
plt.legend()
plt.grid(True)
plt.show()
```
