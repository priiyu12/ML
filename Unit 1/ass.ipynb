```python
import re

def remove_html(text):
    clean_text = re.sub(r'<.*?>', '', text)
    return clean_text
    
def remoce_sc(text):
    clean_text = re.sub(r'[^a-zA-Z0-9\s]', '', text)
    return clean_text

text = "<html>Hello! This is <b>clean</b> text.</html>"

print(remove_html(text))
print(remoce_sc(text))
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Generate skewed data
data = np.random.exponential(scale=2, size=20)

# Apply transformations
log_transformed = np.log1p(data)       # log(1 + x) to avoid log(0)
sqrt_transformed = np.sqrt(data)

plt.figure(figsize=(10, 6))
plt.plot(data, label='Actual', marker='o') 
plt.plot(log_transformed, label='log', marker='+')
plt.plot(sqrt_transformed, label='sqrt', marker='^')
plt.title('Comparison of Original and Transformed Distributions')
plt.xlabel('Index')
plt.ylabel('Value')
plt.legend()
plt.grid(True)
plt.show()

# Plotting
plt.figure(figsize=(15, 5))

# Original data
plt.subplot(1, 3, 1)
plt.hist(data, bins=30, color='skyblue', edgecolor='black')
plt.title('Original Skewed Data')

# Log transformed data
plt.subplot(1, 3, 2)
plt.hist(log_transformed, bins=30, color='lightgreen', edgecolor='black')
plt.title('Log Transformed Data')

# Square root transformed data
plt.subplot(1, 3, 3)
plt.hist(sqrt_transformed, bins=30, color='salmon', edgecolor='black')
plt.title('Square Root Transformed Data')

plt.tight_layout()
plt.show()
```

```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer

# Sample text
text = "The striped bats are hanging on their feet for best."

# Tokenization
tokens = word_tokenize(text)
print("Tokens:", tokens)

# Stemming
stemmer = PorterStemmer()
stemmed_words = [stemmer.stem(word) for word in tokens]
print("Stemmed Words:", stemmed_words)

# Lemmatization
lemmatizer = WordNetLemmatizer()
lemmatized_words = [lemmatizer.lemmatize(word) for word in tokens]
print("Lemmatized Words:", lemmatized_words)
```

```python
from sklearn.preprocessing import MinMaxScaler, OneHotEncoder
import pandas as pd
# Create DataFrame
data = {
    'Income': [15000, 1800, 120000, 10000],
    'Age': [25, 18, 42, 51],
    'Department': ['HR', 'Legal', 'Marketing', 'Management']
}

df = pd.DataFrame(data)

# 1. Scale 'Income' and 'Age' using MinMaxScaler
scaler = MinMaxScaler()
df[['Income', 'Age']] = scaler.fit_transform(df[['Income', 'Age']])
print(df, "\n")


# 2. OneHotEncoder for 'Department'

encoded_df = pd.get_dummies(df, columns=['Department'], dtype='int')

print(encoded_df)
```

```python
# Create a DataFrame
data = {
    'Name': ['Alex', 'Bob', 'Clarke'],
    'Age': [10, 12, 13]
}

df_emp = pd.DataFrame(data)
print(df_emp)

# 2. Read CSV file and display contents
df_csv_emp = pd.read_csv('employees_info.csv')
print(df_csv_emp)

# 3. General info of the DataFrame
df_csv_emp.info()

# 4. Access 'name' and 'gender' columns
print(df_csv_emp[['name', 'gender']])

# 5. Retrieve the first row using .loc[]
print(df_csv_emp.loc[0])

# 6. Get records from row 0 to 5 for specific columns
print(df_csv_emp.loc[0:5, ['name', 'job title']])

# 7. Filter where department is "Accounting"
print(df_csv_emp[df_csv_emp['department'] == 'Accounting'][['name', 'job title', 'department']])

# 8. Delete 'time zone' column
df_csv_emp.drop(columns=['time zone'], inplace=True)

# 9. Drop duplicates
df_csv_emp.drop_duplicates(inplace=True)

# 10. Drop duplicates based on the 'residence' column
df_csv_emp.drop_duplicates(subset=['residence'], inplace=True)

# 11. Drop rows with missing values
df_csv_emp.dropna(inplace=True)

# 12. Drop columns with missing values
df_csv_emp.dropna(axis=1, inplace=True)

# 13. Drop rows/columns with specific thresholds
df_csv_emp.dropna(thresh=2, inplace=True)

# 14. Count missing values in each column
print(df_csv_emp.isnull().sum())

# 15. Calculate percentage of missing values
print(df_csv_emp.isnull().mean() * 100)

# 16. Fill missing values with "Unknown"
df_csv_emp.fillna('Unknown', inplace=True)

# 17. Standardize 'name' column
df_csv_emp['name'] = df_csv_emp['name'].str.title()
df_csv_emp['name'] = df_csv_emp['name'].str.lower()

# 18. Replace gender values
df_csv_emp['gender'] = df_csv_emp['gender'].replace({'M': 'Male', 'F': 'Female'})

# 19. Remove non-numeric characters from 'phone' column
df_csv_emp['phone'] = df_csv_emp['phone'].str.replace(r'\D', '', regex=True)
```
