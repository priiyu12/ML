```python
import re

def remove_html_tags(text):
    clean_text = re.sub(r'<.*?>', '', text)
    return clean_text

def remove_special_characters(text):
    clean_text = re.sub(r'[^a-zA-Z0-9\s]', '', text)
    return clean_text

text = "Hello #GLS @Ahmedabad Studnets <p> <br> at www.glsuniversity.ac.in"
print("Original Text:")
print(text)

remove_html = remove_html_tags(text)
print("\nAfter removing HTML tags:")
print(remove_html)

remove_sChar = remove_special_characters(remove_html)
print("\nAfter removing special characters:")
print(remove_sChar)
```

```python
#Transforming Skewed Distributions
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# DataFrame with Skewed Data
skew_data = {'value': [1, 2, 3, 4, 5, 10, 20, 50, 100, 200, 500, 1000, 2000]}
df = pd.DataFrame(skew_data)

# Visualize original data
plt.plot(df['value'])
plt.title('Original Distribution')
plt.show()

# Apply log transformation
df['log_value'] = np.log(df['value'])

# Visualize log-transformed data
plt.plot(df['log_value'])
plt.title('Log-transformed Distribution')
plt.show()

# Apply square root transformation
df['sqrt_value'] = np.sqrt(df['value'])

# Visualize square root-transformed data
plt.plot(df['sqrt_value'])
plt.title('Square Root-transformed Distribution')
plt.show()
```

```python
#Text Preprocessing

#Tokenization : Splitting text into smaller units
print("Tokenization: \n")

from nltk.tokenize import word_tokenize, sent_tokenize

sentence = 'I am going to a party!. It is going to be  lots of fun. We gonna be eating so much food. I wish it will be a good day!'

print("Word Tokens: ", word_tokenize(sentence))
print("Word Tokens: ", sent_tokenize(sentence))

#Stemming: Reducing words to their root form
print("Stemming: \n")

from nltk.stem.porter import PorterStemmer
stemmer = PorterStemmer()

def stem_words(text):
    return " ".join([stemmer.stem(word) for word in text.split()])

print("Stem Words: ", stem_words(sentence))

#Lemmatization: Converting words to their dictionary form
print("Lemmatization: \n")

from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()

def lemmatize_word(text):
    return ", ". join([lemmatizer.lemmatize(word, pos ='v') for word in text.split()])

print("Lemma Words: [", lemmatize_word(sentence), "]")
```

```python
#Text Preprocessing
import pandas as pd
data = {
    'Text1': ['I love eating pizza', 'Baking is the best thing to do', 'Baking is an art'],
    'Text2': ['I love machine learning', 'Python is great for data', 'Deep learning is powerful']
}
text = pd.DataFrame(data)
print(text)
```

```python
#Tokenization: Splitting text into smaller units (e.g., words, sentences)
from nltk.tokenize import word_tokenize
text['Text1_Token'] = text['Text1'].apply(word_tokenize)
text['Text2_Token'] = text['Text2'].apply(word_tokenize)
print("\nTokenized Text Data: \n", text)
```

```python
#Stemming: Reducing words to their root form (e.g., running  run).
from nltk.stem import PorterStemmer
stemmer = PorterStemmer()

text['Text1_Stem'] = text['Text1_Token'].apply(lambda tokens: [stemmer.stem(word) for word in tokens])
text['Text2_Stem'] = text['Text2_Token'].apply(lambda tokens: [stemmer.stem(word) for word in tokens])

print("\nStemmed Text Data: \n")
print(text[['Text1', 'Text1_Stem']], "\n")
print(text[['Text2', 'Text2_Stem']])
```

```python
#Lemmatization: Converting words to their dictionary form (e.g., better  good)
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()

text['Text1_Lemma'] = text['Text1_Token'].apply(lambda tokens: [lemmatizer.lemmatize(word, pos='v') for word in tokens])
text['Text2_Lemma'] = text['Text2_Token'].apply(lambda tokens: [lemmatizer.lemmatize(word, pos='v') for word in tokens])
print("\nLemmatized Text Data: \n")
print(text[['Text1', 'Text1_Lemma']], "\n")
print(text[['Text2', 'Text2_Lemma']])
```

```python
import pandas as pd

data = {'timestamp': ['2025-01-06 08:23:45', '2025-06-15 14:12:00', '2025-12-25 18:35:00']}
convert = pd.DataFrame(data)
print(convert, "\n")

# Convert to datetime
convert['timestamp'] = pd.to_datetime(convert['timestamp'])

# Extract features
convert['day_of_week'] = convert['timestamp'].dt.dayofweek  # Extract day of the week (0=Monday, 6=Sunday)

convert['hour_of_day'] = convert['timestamp'].dt.hour  # Extract hour of the day (0 to 23)

convert['month'] = convert['timestamp'].dt.month  # Extract month of the year (1 to 12)

convert['day_of_month'] = convert['timestamp'].dt.day  # Extract day of the month (1 to 31)

convert['quarter'] = convert['timestamp'].dt.quarter  # Extract quarter of the year (1 to 4)

convert['is_weekend'] = convert['timestamp'].dt.dayofweek >= 5  # Extract whether it's a weekend or weekday (1 for weekend, 0 for weekday)

def get_season(month):
    if month in [12, 1, 2]:
        return 'Winter'
    elif month in [3, 4, 5]:
        return 'Spring'
    elif month in [6, 7, 8]:
        return 'Summer'
    else:
        return 'Fall'

convert['season'] = convert['month'].apply(get_season)
print(convert, "\n")
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# DataFrame with skewed data
skew_data = {'value': [1, 2, 3, 4, 5, 10, 20, 50, 100, 200, 400, 500, 1000, 2000]}
df = pd.DataFrame(skew_data)

# Apply transformations
df['log_value'] = np.log(df['value'] + 1)  # Add 1 to handle potential zeros
df['sqrt_value'] = np.sqrt(df['value'])

# Combined plot
plt.figure(figsize=(10, 6))
plt.plot(df['value'], label='Original', marker='o')
plt.plot(df['log_value'], label='Log-transformed', marker='s')
plt.plot(df['sqrt_value'], label='Square Root-transformed', marker='^')
plt.title('Comparison of Original and Transformed Distributions')
plt.xlabel('Index')
plt.ylabel('Value')
plt.legend()
plt.grid(True)
plt.show()

# Histogram for distributions
plt.figure(figsize=(15, 5))

# Original
plt.subplot(1, 3, 1)
plt.hist(df['value'], bins=10, color='blue', alpha=0.7, edgecolor='black')
plt.title('Original Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')

# Log-transformed
plt.subplot(1, 3, 2)
plt.hist(df['log_value'], bins=10, color='orange', alpha=0.7, edgecolor='black')
plt.title('Log-transformed Distribution')
plt.xlabel('Log(Value)')

# Square Root-transformed
plt.subplot(1, 3, 3)
plt.hist(df['sqrt_value'], bins=10, color='green', alpha=0.7, edgecolor='black')
plt.title('Square Root-transformed Distribution')
plt.xlabel('Sqrt(Value)')

plt.tight_layout()
plt.show()
```
