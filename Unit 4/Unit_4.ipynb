```python
#K-MEAN CLUSTERING WITH ARRAYS
import numpy as np  # For numerical computations and array handling
import matplotlib.pyplot as plt  # For visualization (scatter plot)
from sklearn.cluster import KMeans  # Importing K-Means clustering from sklearn

data = np.array([
    [180, 44], [175, 76], [189, 87], [165, 87], [170, 60],
    [160, 55], [160, 90], [179, 80], [160, 82], [171, 52],
    [173, 63], [182, 78], [177, 89], [167, 58], [150, 55]
])

kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
kmeans.fit(data)  # Perform clustering on the dataset

labels = kmeans.labels_  # Get the cluster index (0 or 1) assigned to each data point
centroids = kmeans.cluster_centers_  # Get the coordinates of cluster centers
print("Cluster Labels:", labels)  # Print which data point belongs to which cluster

plt.figure(figsize=(8, 6))
plt.scatter(
    data[:, 0], data[:, 1],  # X-axis: Height, Y-axis: Weight
    c=labels,  
    cmap='viridis',# Color points based on cluster label
    marker='o', s=70,             # Circle marker for data points
    label="Data Points"
)
plt.scatter(
    centroids[:, 0], centroids[:, 1],  # X-axis: Height, Y-axis: Weight
    c='red', 
    marker='^', s=100,  # Red 'X' marker with size 200
    label="Centroids"
)

plt.title("K-Means Clustering (K=3) - List/Array")  # Title of the plot
plt.xlabel("Height (cm)")  # Label for x-axis
plt.ylabel("Weight (kg)")  # Label for y-axis
plt.legend()  # Show legend (Data Points & Centroids)
plt.grid(True)  # Enable grid for better readability
plt.show()  # Display the scatter plot
```

```python
#K-MEAN CLUSTERING WITH CSV
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans  # K-Means algorithm from Scikit-Learn

data = pd.read_csv("data.csv")
X = data[['Height', 'Weight']].values  # Convert to NumPy array

kmeans = KMeans(
    n_clusters=2, 
    random_state=42, 
    n_init=10
)
kmeans.fit(X)  # Train the model on our data

labels = kmeans.labels_  # Labels assigned to each data point
centroids = kmeans.cluster_centers_  # Coordinates of cluster centers
print("\nCluster Labels:", labels)

plt.figure(figsize=(10,6)) # Set figure size
plt.scatter(
    X[:, 0], X[:, 1],
    c=labels,
    cmap='viridis',
    marker='o',
    s=70, label="Data Points"
)
plt.scatter(
    centroids[:, 0], centroids[:, 1],
    c='red',
    marker='^',
    s=100, label="Centroids"
)

plt.title("K-Means Clustering (K=2) - CSV File")  # Plot title
plt.xlabel("Height")  # X-axis label
plt.ylabel("Weight")  # Y-axis label
plt.legend()  # Show legend
plt.grid(True)  # Show grid
plt.show()  # Display the plot
```
