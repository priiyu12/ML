```python
import pandas as pd
import matplotlib.pyplot as plt

# Create a dataset for customer ages
ages = [22, 25, 29, 30, 35, 38, 40, 42, 45, 50]
df = pd.DataFrame({'Age': ages})

# Plot histogram
plt.hist(df['Age'], bins=5, edgecolor='black', color='blue')
plt.title('Age distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Age data with missing values
data = {'Age': [25, 30, 35, None, 28, 32, 45, None, 30, 28]}
df = pd.DataFrame(data)

# Remove missing values
df_cleaned = df.dropna()

# Plot histogram
plt.hist(df_cleaned['Age'], bins=5, edgecolor='black', color='blue')
plt.title('Age distribution after removing missing values')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data for shapes
shapes = ['Square1', 'Rect', 'Square2']
lengths = [10, 15, 20]
heights = [12, 18, 25]

# Group Bar Graph
x = np.arange(len(shapes))
width = 0.35

fig, ax = plt.subplots()
ax.bar(x - width / 2, lengths, width, label='Length', color='blue')
ax.bar(x + width / 2, heights, width, label='Height', color='red')

ax.set_ylabel('Values')
ax.set_title('Group Bar Graph for Shapes')
ax.set_xticks(x)
ax.set_xticklabels(shapes)
ax.legend()

plt.show()

# Stacked Bar Graph
fig, ax = plt.subplots()
ax.bar(shapes, lengths, label='Length', color='blue')
ax.bar(shapes, heights, bottom=lengths, label='Height', color='red')

ax.set_ylabel('Values')
ax.set_title('Stacked Bar Graph for Shapes')
ax.legend()

plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
transaction_amounts = [10, 15, 20, 25, 30, 35, 1000, 2000]
cv_scores_model_1 = [0.82, 0.84, 0.83, 0.81, 0.95]
cv_scores_model_2 = [0.88, 0.87, 0.89, 0.86, 0.88]
daily_returns = [0.01, 0.02, 0.015, -0.03, 0.02, -0.1, 0.03, 0.08]

# Box plot
plt.boxplot([transaction_amounts, cv_scores_model_1, cv_scores_model_2, daily_returns], labels=['Transaction Amounts', 'Model 1 CV', 'Model 2 CV', 'Daily Returns'])
plt.title('Box Plot for Various Data')
plt.show()
```

```python
import seaborn as sns
import numpy as np
import pandas as pd

# Data for heatmap
data = {
    'Feature1': np.random.rand(100),
    'Feature2': np.random.rand(100) * 2,
    'Feature3': np.random.rand(100) * 3,
    'Target': np.random.rand(100) * 4,
}

df = pd.DataFrame(data)

plt.figure(figsize=(8, 6))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Heatmap of Feature Correlations')
plt.show()

# True vs Predicted values heatmap
true_values = np.random.rand(10)
predicted_values = true_values + np.random.normal(0, 0.1, 10)

plt.figure(figsize=(8, 6))
sns.heatmap(np.corrcoef(true_values, predicted_values), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Heatmap of True vs Predicted Values')
plt.show()
```
