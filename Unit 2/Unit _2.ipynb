```python
#BarChart (MatPlotLib) : Horix
import matplotlib.pyplot as plt

# Data
categories = ['A', 'B', 'C', 'D']
values = [10, 20, 15, 25]

plt.barh(categories, values, color='rosybrown')

plt.xlabel('Values')
plt.ylabel('Categories')
plt.title('Horizontal Bar Chart')
plt.show()
```

```python
#BarChart (MatPlotLib) : Grouped
import numpy as np
import matplotlib.pyplot as plt

categories = ['A', 'B', 'C']
values1 = [10, 15, 20]
values2 = [12, 18, 25]

x = np.arange(len(categories))
width = 0.35

plt.bar(x - width/2, values1, width, label='Group 1', color='seagreen')
plt.bar(x + width/2, values2, width, label='Group 2', color='lawngreen')

plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Grouped Bar Chart')
plt.xticks(x, categories)
plt.legend()
plt.show()
```

```python
#BarChart (MatPlotLib) : Stacked
import matplotlib.pyplot as plt

categories = ['A', 'B', 'C']
values1 = [10, 15, 20]
values2 = [5, 10, 15]

plt.bar(categories, values1, label='Group 1', color='hotpink')
plt.bar(categories, values2, label='Group 2', color='mediumpurple', bottom=values1)

plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Stacked Bar Chart')
plt.legend()
plt.show()
```

```python
#BoxPlot (Seaborn) : With Outliers
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

data = data = [10, 34, 44, 56, 32, 77, 41, 99, 31]
df = pd.DataFrame({'Values': data})

sns.boxplot(data=df, y='Values', color='hotpink')

plt.title('Box Plot with Outlier Detection')

Q1 = df['Values'].quantile(0.25)
Q3 = df['Values'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
outliers = df[(df['Values'] < lower_bound) | (df['Values'] > upper_bound)]
print(f"Outliers:\n{outliers}")
plt.show()
```

```python
#Heatmap (Seaborn) : Correlation Matrix
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

np.random.seed(42)
data = {
    'Feature_1': np.random.normal(50, 10, 100),
    'Feature_2': np.random.normal(30, 5, 100),
    'Feature_3': np.random.normal(10, 2, 100),
    'Feature_4': np.random.uniform(20, 40, 100)
}
df = pd.DataFrame(data)

correlation_matrix = df.corr()

sns.heatmap(correlation_matrix, annot=True, cmap='Blues')

plt.title('Correlation Matrix Heatmap')
plt.show()
```
